```r
library(rpart)
library(rattle)
str(kyphosis)
head(kyphosis,5)
```

```r

tree <- rpart(formula = Kyphosis ~.,data = kyphosis, method = 'class')
printcp(tree)
# older way to plot
#plot(tree,uniform = T,main = 'Kyphosis Tree')
#text(tree,use.n = T,all = T)
# good option is to use rpart.plot
library(rpart.plot)
prp(tree)
#help(rpart)
```

```r
## or better way to plot above is 
library(rpart)
library(rattle)
depFit <- rpart(data=kyphosis,formula = Kyphosis ~., method = 'class') ##xval=3)
fancyRpartPlot(depFit)
printcp(depFit)
```
